# Data Cleaning

```python
missing_values = df.isnull().sum()
unique_values = df.nunique()
outliers = {}
```

```python
# Check for obvious outliers in numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
for column in numerical_columns:
    # Using IQR to detect outliers
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outlier_condition = ((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))
    outliers[column] = df[outlier_condition][column].count()
```

```python
data_cleaning_info = {
    "Missing Values": missing_values,
    "Unique Values per Column": unique_values,
    "Potential Outliers in Numerical Columns": outliers
}
```

```python
data_cleaning_info
```

```python
# Handling Missing Values
# Fill missing numerical values with the median (to avoid the influence of outliers)
# and categorical values with the mode (most frequent value)
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].median(), inplace=True)
```

```python
missing_values_after_cleaning = df.isnull().sum()
```

```python
cleaning_summary = {
    "Missing Values Before Cleaning": missing_values,
    "Missing Values After Cleaning": missing_values_after_cleaning
}

cleaning_summary
```

```python
def remove_non_numeric_chars(series):
    return series.replace('[^\d.]', '', regex=True)
```

```python
df['Monthly income before tax?'] = remove_non_numeric_chars(df['Monthly income before tax?'])
```

```python
# Replace empty strings with NaN
df['Monthly income before tax?'] = df['Monthly income before tax?'].replace('', np.nan)
```

```python
# Convert the column to numeric, coercing errors
df['Monthly income before tax?'] = pd.to_numeric(df['Monthly income before tax?'], errors='coerce')
```

```python
# Now calculate the median income by occupation
median_income_by_occupation = df.groupby('Occupation')['Monthly income before tax?'].median()
```

```python
# handeling missing values in 'Monthly income before tax?' based on the median of their respective 'Occupation' group

for occupation in median_income_by_occupation.index:
    median_value = median_income_by_occupation[occupation]
    filter_condition = (df['Occupation'] == occupation) & (df['Monthly income before tax?'].isna())
    df.loc[filter_condition, 'Monthly income before tax?'] = median_value
```

```python
df.head(20)
```

```python
# Replace non-numeric values like 'Vet ej' with NaN
df['Annual electricity consumption'] = df['Annual electricity consumption'].replace('Vet ej', np.nan)
```

```python
# Removing units 
df['Annual electricity consumption'] = df['Annual electricity consumption'].str.replace(' kwh', '', regex=False)
```

```python
df['Annual electricity consumption'] = pd.to_numeric(df['Annual electricity consumption'], errors='coerce')
```

```python
df.head()
```

```python
median_electricity_consumption = df['Annual electricity consumption'].median()
```

```python
# Imputing the missing values
df['Annual electricity consumption'].fillna(median_electricity_consumption, inplace=True)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Creating dummy variables for categorical columns

occupation_dummies = pd.get_dummies(df['Occupation'], drop_first=True)
area_dummies = pd.get_dummies(df['Area'], drop_first=True)
age_group_dummies = pd.get_dummies(df['How old are you?'], drop_first=True)
housing_dummies = pd.get_dummies(df['What type of housing do you live in?'], drop_first=True)
```

```python
# Preparing the data for the regression model

X = pd.concat([df[['Monthly income before tax?']], occupation_dummies, area_dummies, age_group_dummies, housing_dummies], axis=1)
y = df['Annual electricity consumption']
```

```python
# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# linear regression model

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R²): {r2}")
print("\nModel Coefficients:")
print(pd.Series(model.coef_, index=X.columns).sort_values())
```
